```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import random as rd
```

La legge dei grandi numeri è un teorema fondamentale della teoria della probabilità che indica che se ripetiamo molte volte (tendendo all'infinito) lo stesso esperimento, la frequenza di un certo evento tende ad essere costante.

Vale a dire, la legge dei grandi numeri indica che se lo stesso test viene eseguito ripetutamente (ad esempio, lancio di una moneta), la frequenza con cui si ripeterà un determinato evento (testa/croce) si avvicinerà a una costante. Questa a sua volta sarà la probabilità che questo evento si verifichi.

1. Scrivere una funzione che simula il lancio di una moneta per N volte e restituisce la frequenza % di volte che è uscita testa.

```python
def simulazione_lanci(N):
    lanci = [np.random.randint(0, 2) for i in range(N)]
    return (sum(lanci) / N ) * 100

simulazione_lanci(100000)
```

2. Calcolare la frequenza % con la grandezze del campione (numero di lanci) che varia 100 volte da 10 a 20000. Ottengo quindi 100 valori

```python
# Calcolo l'array di campioni
campioni = [n for n in range(10, 20000, 200)]
campioni
```

```python
# Ottengo il valore per ogni campione calcolato
frequenza_campioni = [simulazione_lanci(n) for n in campioni]
frequenza_campioni
```

```python
# Creo il DataFrame
df = pd.DataFrame()
df['Campione'] = campioni
df['Frequenza'] = frequenza_campioni
df
```

2. Fare il plot del Numero medio di teste (y) rispetto alla grandezza del campione (x)

```python
# Creo il plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x=df['Campione'], y=df['Frequenza'])
plt.axhline(50, color='red', linestyle='--')
plt.xlabel('Numero di lanci')
plt.ylabel('Frequenza testa')
plt.title('Legge dei grandi numeri')
plt.show()
```
