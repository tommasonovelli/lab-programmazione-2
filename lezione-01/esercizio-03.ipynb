1. Creare un array a monodimensionale 

```python
a = np.arange(1, 5)
a
```

2. Genero un nuovo array b, partendo da a

```python
b = a[1:4]
b
```

3. Genero un nuovo array c dato dall'inversione di a

```python
c = a[::-1]
c
```

4. Divido l'array a per l'array c

```python
a / c
```

5. Faccio le stesse cose solo usando le liste di python

```python
lista_a = list(range(1, 5))
lista_a
```

```python
lista_b = lista_a[1:4]
lista_b
```

```python
lista_c = lista_a[::-1]
lista_c
```

```python
# lista_a / lista_c genera un TypeError
[lista_a[i] / lista_c[i] for i in range(len(lista_a))]
```
