```python
# Import Package
import numpy as np
```

1. Creare un vettore di numeri primi compresi tra 0 e 10

```python
numeri_primi = np.array([2, 3, 5, 7])
```

2. Conta quanti numeri ci sono nel vettore

```python
# Usando len
print(len(numeri_primi))

# Usando .size 
print(numeri_primi.size)
```

3. Quale pensi sia il tipo di dato dtype del vettore?
Il dtype è di tipo intero

```python
print(numeri_primi.dtype)
```

4. Scrivi l'array usando una list comprehension che controlla che i numeri siano primi

```python
lista_primi = np.array([numero for numero in range(10) if numero in numeri_primi])
lista_primi
```
