## 1. Trasformare cicli in list comprehension

Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension

### Esempio 1
Prendo solo i nomi che cominciano con la m

```python
lista_nomi = ["marco", "mario", "luca", "luigi"]
```

```python
# ciclo
nomi_con_m = []

for nome in lista_nomi:
    if nome.startswith("m"):
        nomi_con_m.append(nome)

print(nomi_con_m)
```

```python
# list comprehension
nomi_con_m = [nome for nome in lista_nomi if nome.startswith("m")]

print(nomi_con_m)
```

### Esempio 2
Prendo solo i numeri che sono maggiori di 5

```python
numeri = [2, 7, 4, 9, 1, 6]
```

```python
# ciclo 
maggiori_di_cinque = []

for numero in numeri:
    if numero > 5:
        maggiori_di_cinque.append(numero)

print(maggiori_di_cinque)
```

```python
# list comprehension
maggiori_di_cinque = [numero for numero in numeri if numero > 5]

print(maggiori_di_cinque)
```

### Esempio 3
Potenza terza dei numeri

```python
numeri = [1, 2, 3, 4, 5]
```

```python
# ciclo
potenze_terze = []

for numero in numeri:
    potenze_terze.append(numero ** 3)

print(potenze_terze)
```

```python
# list comprehension
potenze_terze = [numero ** 3 for numero in numeri]

print(potenze_terze)
```
