```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import math

temp_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
```

```python
# Definisco le funzioni per calcolare MAE e RMSE
def calculate_rmse(y_real, y_pred):
    if len(y_real) != len(y_pred):
        raise ValueError("Le liste devono avere la stessa lunghezza.")
    mse = sum((y_real[i] - y_pred[i]) ** 2 for i in range(len(y_real))) / len(y_real)
    return math.sqrt(mse)

def calculate_mae(y_real, y_pred):
    if len(y_real) != len(y_pred):
        raise ValueError("Le liste devono avere la stessa lunghezza.")
    errors = [abs(y_real[i] - y_pred[i]) for i in range(len(y_real))]
    return sum(errors) / len(errors)
```

Prova a fittare i dati sottostanti con le migliori curve, calcola il MAE e l'RMSE

```python
# Definisco la coordinata x e y per temp_max e temp_min
x1 = np.arange(0, len(temp_max))
y1 = temp_max

x2 = np.arange(0, len(temp_min))
y2 = temp_min

# Definisco il modello sinusoidale
def sinusoid(x, a, b, c, d):
    return a * np.sin(b * x + c) + d

# Calcolo i punti di partenza
p0_1 = [(y1.max() - y1.min()) / 2, 2*np.pi/12, 0, y1.mean()]
p0_2 = [(y2.max() - y2.min()) / 2, 2*np.pi/12, 0, y2.mean()]

# Faccio il fit con scipy di temp_max
params1, _ = sp.optimize.curve_fit(sinusoid, x1, y1, p0=p0_1)
a1, b1, c1, d1 = params1

# Faccio il fit con scipy di temp_min
params2, _ = sp.optimize.curve_fit(sinusoid, x2, y2, p0=p0_2)
a2, b2, c2, d2 = params2

# Output dei parametri
print(f"Parametro a: {a1}, {a2}")
print(f"Parametro b: {b1}, {b2}")
print(f"Parametro c: {c1}, {c2}")
print(f"Parametro d: {d1}, {d2}")

# Predizione
y_fit_1 = sinusoid(x1, a1, b1, c1, d1)
y_fit_2 = sinusoid(x2, a2, b2, c2, d2)

# Plot
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].scatter(x1, y1, color='red')
ax[0].plot(x1, y_fit_1, color='blue')
ax[0].set_title('Temp Max')

ax[1].scatter(x2, y2, color='red')
ax[1].plot(x2, y_fit_2, color='blue')
ax[1].set_title('Temp Min')
```

```python
# Calcolo ora RMSE e MAE di temp_min e temp_max

mae_1 = calculate_mae(y1, y_fit_1)
mae_2 = calculate_mae(y2, y_fit_2)

rmse_1 = calculate_rmse(y1, y_fit_1)
rmse_2 = calculate_rmse(y2, y_fit_2)

print(f'MAE MAX: {mae_1}')
print(f'MAE MIN: {mae_2}')
print(f'RMSE MAX: {rmse_1}')
print(f'RMSE MIN: {rmse_2}')
```
