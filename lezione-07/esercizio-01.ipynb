```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

1. Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1. Suggerimento: il parametro di forma viene passato come primo argomento quando si "congela" la distribuzione

```python
# Genero un campione di 1000 valori da una distribuzione gamma
dist = sp.stats.gamma(a=1)
sample = dist.rvs(size=1000)
```

2. Traccia l’istogramma del campione e sovrapponi la PDF della distribuzione.

```python
# Plot
plt.hist(sample, bins=100, density=True) 
x = np.linspace(0, 10)
plt.plot(x, dist.pdf(x), label='PDF')
plt.show()
```

3. Stima il parametro di forma dal campione usando il metodo fit

```python
a, loc, scale = sp.stats.gamma.fit(sample)
a
```

4. Traccia la funzione di distribuzione cumulativa (CDF)

```python
plt.plot(x, dist.cdf(x), label='CDF')
plt.show()
```

5. Calcola la varianza

```python
dist.var()
```
