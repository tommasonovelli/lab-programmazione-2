```python
import numpy as np
```

1. Carica tramite numpy il file generato e salvato a lezione sulle emissioni di CO2

```python
emissions_file = './data/emissioni.txt'
co2_emissions = np.loadtxt(emissions_file, delimiter=None, dtype=float)
co2_emissions
```

2. Calcola quanti valori ci CO2 sono sopra la soglia di 18 tonnellate per anno ed il loro valore medio

```python
len(co2_emissions[co2_emissions > 18])
```

3. Calcola la frazione totale di emissione sopra la soglia (somma totale sopra la soglia/somma totale)

```python
sum_over_threshold = np.sum(co2_emissions[co2_emissions > 18])
total_sum = np.sum(co2_emissions)

# Frazione
sum_over_threshold / total_sum
```
