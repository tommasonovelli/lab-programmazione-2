## 3. Seleziona il numero più vicino a 0.5 in un array casuale

Genera un array 10 x 3 con numeri casuali nell'intervallo [0,1]. Per ogni riga, seleziona il numero più vicino a 0.5. Per farlo usa il fancy indexing.

```python
%conda install numpy
```

```python
import numpy as np
```

Creare un array 10 x 3 con numeri casuali all'interno dell'intervallo [0, 1]

```python
a = np.random.rand(10, 3)
a
```

Per ogni riga, seleziona il numero più vicino a 0.5

```python
dist = np.abs(a - 0.5)              # Calcolo la distanza di ogni valore da 0.5

colonne = np.argmin(dist, axis=1)   # Ottengo l'indice del valore più vicino a 0.5 per ogni riga
righe = np.arange(a.shape[0])       # Ottengo il numero di righe della matrice   

a[righe, colonne]
```
